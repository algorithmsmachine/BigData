```python
from pyspark import SparkContext, SparkConf
from pyspark.sql import SQLContext
import sys

conf = SparkConf().setAppName("Log Analyzer")
sc = SparkContext(conf=conf)
```

```python
filepath= "C:\\Users\\abisht\\FS_logs\\freeswitchlogs1.log"
sc.textFile(filepath).keyBy(lambda  line:   line.split(' ')[2]).take(10)
```

```python
sc.textFile(filepath).keyBy(lambda  line: line.split(' '))\
.map(lambda fields: (fields[0], (fields[1], fields[2])))\
.take(10)
```

# Pyspark 

```python
import pyspark.pandas as ps
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
rdd = spark.sparkContext.textFile(filepath)
# print(rdd.collect())
```

```python
print("Count : "+str(rdd.count()))
```

```python
rdd2 = rdd.flatMap(lambda x: x.split(" "))
# print(rdd2.collect())
```

```python
rdd3 = rdd2.map(lambda x: (x,1))
print(rdd3.collect())
```

```python
rdd4 = rdd.filter(lambda x : 'INVITE' in x[2])
print(rdd4.collect())
```

```python
rdd5 = rdd4.reduceByKey(lambda a,b: a+b)
print(rdd5.collect())
```

```python
rdd6 = rdd5.map(lambda x: (x[1],x[0])).sortByKey()
print(rdd6.collect())
```

```python
rdd6.saveAsTextFile("Call_INVITE_count")
```
